- **Stochastic Gradient Descent (SGD)**:
  An iterative optimization algorithm used for minimizing loss functions in linear classifiers and regressors, particularly useful for large datasets.

- **Decision Trees**:
  A tree-based algorithm where each internal node represents a decision on a feature, each leaf node represents an outcome, and paths from root to leaf represent classification rules.

- **K-Nearest Neighbors (KNN)**:
  A lazy learning algorithm that classifies a data point based on the majority class of its k-nearest neighbors in the feature space.
  
- **Naive Bayes (NB)**:
  A probabilistic classifier based on Bayes' theorem, assuming strong independence between features.

- **Bagging (Bootstrap Aggregating)**:

```python
def chi_square_test_of_homogeneity(df, col1, col2):
    """
    Performs a chi-square test of homogeneity on two categorical columns in a DataFrame.

    Args:
        df (pd.DataFrame): The input DataFrame.
        col1 (str): The name of the first categorical column.
        col2 (str): The name of the second categorical column.

    Returns:
        tuple: A tuple containing the chi-square statistic, the p-value, degrees of freedom, and expected frequencies.
    """

    # Create a contingency table
    contingency_table = pd.crosstab(df[col1], df[col2])

    # Perform the chi-square test of homogeneity
    chi2, p, dof, expected = stats.chi2_contingency(contingency_table)

    return chi2, p, dof, expected

# Example usage with the provided data in the user's notebook
# Assume sd48_sample_encoded is the DataFrame with encoded categorical data
if 'sd48_sample_encoded' in locals():
  # Check if the columns exist in the DataFrame
  if 'Import_Export' in sd48_sample_encoded.columns and 'Category' in sd48_sample_encoded.columns:
    # Run the test of homogeneity on 'Import_Export' and 'Category'
    chi2, p, dof, expected = chi_square_test_of_homogeneity(sd48_sample_encoded, 'Import_Export', 'Category')

    # Print the results
    print(f"Chi-square statistic: {chi2}")
    print(f"P-value: {p}")
    print(f"Degrees of freedom: {dof}")
    print("Expected frequencies:")
    print(expected)

    # Interpret the results
    alpha = 0.05
    print(f"\nSignificance level: {alpha}")
    if p < alpha:
        print("The p-value is less than the significance level.")
        print("Reject the null hypothesis: there is evidence of a difference in distribution between the categories of the two variables.")
    else:
        print("The p-value is greater than the significance level.")
        print("Fail to reject the null hypothesis: there is not enough evidence to suggest a difference in distribution between the categories of the two variables.")

  else:
    print("Columns 'Import_Export' or 'Category' not found in sd48_sample_encoded.")
else:
    print("The DataFrame 'sd48_sample_encoded' does not exist.")
```

```python
# Test of Normality {Shapiro-Wilk | Kolmogorov-Smirnov | Anderson-Darling | Jarque-Bera}

def normality_tests(data):

    sd48_results = {}

    # Shapiro-Wilk test
    shapiro_statistic, shapiro_pvalue = shapiro(data)
    sd48_results["Shapiro-Wilk"] = {"statistic": shapiro_statistic, "p-value": shapiro_pvalue}

    # Kolmogorov-Smirnov test (against a normal distribution)
    ks_statistic, ks_pvalue = kstest(data, 'norm')
    sd48_results["Kolmogorov-Smirnov"] = {"statistic": ks_statistic, "p-value": ks_pvalue}

    # Anderson-Darling test
    anderson_result = anderson(data)
    sd48_results["Anderson-Darling"] = {
        "statistic": anderson_result.statistic,
        "critical_values": anderson_result.critical_values,
        "significance_level": anderson_result.significance_level
    }

    # Jarque-Bera test
    jb_statistic, jb_pvalue = jarque_bera(data)
    sd48_results["Jarque-Bera"] = {"statistic": jb_statistic, "p-value": jb_pvalue}

    return sd48_results
```

```python
# Test of Correlation (T-test)
def t_test_correlation(data1, data2):
    t_statistic, p_value = stats.ttest_ind(data1, data2)
    return {"t-statistic": t_statistic, "p-value": p_value}

for col in sd48_sample_encoded[non_categorical_cols]:
    print(f"Normality tests for {col}:")
    normality_test_results = normality_tests(sd48_sample_encoded[col])
    for test_name, result in normality_test_results.items():
        print(f"  {test_name}: {result}")

    # Example of T-test
    if len(non_categorical_cols) > 1:
        for i in range(len(non_categorical_cols)):
            for j in range(i+1, len(non_categorical_cols)):
                col1, col2 = non_categorical_cols[i], non_categorical_cols[j]
                t_test_result = t_test_correlation(sd48_sample_encoded[col1], sd48_sample_encoded[col2])
                print(f"T-test Correlation between {col1} and {col2}: {t_test_result}")
    print()
    print('*'*20)
    print()
```

```python
import time
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

def time_model_run(model, X_train, y_train):
    start_time = time.time()
    model.fit(X_train, y_train)
    end_time = time.time()
    return end_time - start_time

# Logistic Regression
log_reg_model = LogisticRegression(random_state=55022)
log_reg_time = time_model_run(log_reg_model, sd48_X_train, sd48_y_train)
print(f"Logistic Regression runtime: {log_reg_time:.4f} seconds")

# Support Vector Machine (SVM)
svm_model = SVC(random_state=55022)
svm_time = time_model_run(svm_model, sd48_X_train, sd48_y_train)
print(f"SVM runtime: {svm_time:.4f} seconds")

# Stochastic Gradient Descent (SGD)
sgd_model = SGDClassifier(random_state=55022)
sgd_time = time_model_run(sgd_model, sd48_X_train, sd48_y_train)
print(f"SGD runtime: {sgd_time:.4f} seconds")

# Decision Tree
dt_model = DecisionTreeClassifier(random_state=55022)
dt_time = time_model_run(dt_model, sd48_X_train, sd48_y_train)
print(f"Decision Tree runtime: {dt_time:.4f} seconds")

# K-Nearest Neighbors (KNN)
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_time = time_model_run(knn_model, sd48_X_train, sd48_y_train)
print(f"KNN runtime: {knn_time:.4f} seconds")

# Naive Bayes
nb_model = GaussianNB()
nb_time = time_model_run(nb_model, sd48_X_train, sd48_y_train)
print(f"Naive Bayes runtime: {nb_time:.4f} seconds")

# Random Forest
rf_model = RandomForestClassifier(random_state=55022)
rf_time = time_model_run(rf_model, sd48_X_train, sd48_y_train)
print(f"Random Forest runtime: {rf_time:.4f} seconds")

# Extreme Gradient Boosting (XGBoost)
xgb_model = XGBClassifier(random_state=55022)
xgb_time = time_model_run(xgb_model, sd48_X_train, sd48_y_train)
print(f"XGBoost runtime: {xgb_time:.4f} seconds")
```

```python
import tracemalloc
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

def get_model_memory_usage(model, X, y):
    tracemalloc.start()
    model.fit(X, y)
    current, peak = tracemalloc.get_traced_memory()
    tracemalloc.stop()
    return peak / (1024 * 1024)  # Convert bytes to MB

# Logistic Regression
log_reg_model = LogisticRegression(random_state=55022)
log_reg_memory = get_model_memory_usage(log_reg_model, sd48_X_train, sd48_y_train)
print(f"Logistic Regression memory usage: {log_reg_memory:.2f} MB")

# Support Vector Machine (SVM)
svm_model = SVC(random_state=55022)
svm_memory = get_model_memory_usage(svm_model, sd48_X_train, sd48_y_train)
print(f"SVM memory usage: {svm_memory:.2f} MB")

# Stochastic Gradient Descent (SGD)
sgd_model = SGDClassifier(random_state=55022)
sgd_memory = get_model_memory_usage(sgd_model, sd48_X_train, sd48_y_train)
print(f"SGD memory usage: {sgd_memory:.2f} MB")

# Decision Tree
dt_model = DecisionTreeClassifier(random_state=55022)
dt_memory = get_model_memory_usage(dt_model, sd48_X_train, sd48_y_train)
print(f"Decision Tree memory usage: {dt_memory:.2f} MB")

# K-Nearest Neighbors (KNN)
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_memory = get_model_memory_usage(knn_model, sd48_X_train, sd48_y_train)
print(f"KNN memory usage: {knn_memory:.2f} MB")

# Naive Bayes
nb_model = GaussianNB()
nb_memory = get_model_memory_usage(nb_model, sd48_X_train, sd48_y_train)
print(f"Naive Bayes memory usage: {nb_memory:.2f} MB")

# Random Forest
rf_model = RandomForestClassifier(random_state=55022)
rf_memory = get_model_memory_usage(rf_model, sd48_X_train, sd48_y_train)
print(f"Random Forest memory usage: {rf_memory:.2f} MB")

# Extreme Gradient Boosting (XGBoost)
xgb_model = XGBClassifier(random_state=55022)
xgb_memory = get_model_memory_usage(xgb_model, sd48_X_train, sd48_y_train)
print(f"XGBoost memory usage: {xgb_memory:.2f} MB")
```
