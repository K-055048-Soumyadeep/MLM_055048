# Unsupervised Learning for International Trade Transactions

## Project Overview

This project explores unsupervised machine learning techniques, specifically clustering, to segment and analyze a dataset of international trade transactions. The primary goal is to identify distinct groups or segments within the data based on their similarities and derive meaningful insights for business applications.

## Data Source

The dataset used in this project is sourced from Kaggle:

- Dataset Name: Imports-Exports-15000
- Size: 2.69 MB

## Data Description

The dataset contains information about international trade transactions, including details about products, quantities, values, countries, and other relevant attributes.

- Data Type: Panel
- Data Dimension: 16 Variables, 15000 Observations
- Data Variable Types: Numeric and Non-Numeric
- Data Variable Categories: Index, Nominal, Ordinal, Non-Categorical

## Project Objectives

- *Segmentation:* Segment the dataset using unsupervised machine learning clustering algorithms.
- *Cluster Identification:* Identify the appropriate number of segments or clusters.
- *Cluster Characterization:* Determine the characteristics of each segment or cluster.

## Methodology

The project utilizes the following clustering algorithms:

- *K-Means:* A partitioning-based clustering algorithm that aims to group data points into k clusters based on their similarity to cluster centroids.
- *DBSCAN:* A density-based clustering algorithm that identifies clusters based on the density of data points in a region.

*Steps:*

1. *Data Preprocessing:* Relevant variables are selected, and the data is scaled using MinMaxScaler and encoded using Ordinal Encoder.
2. *Clustering:* K-Means and DBSCAN algorithms are applied with different parameters.
3. *Cluster Evaluation:* The Silhouette Score and Davies-Bouldin Score are calculated to assess the quality of clusters.
4. *Cluster Selection:* The optimal number of clusters is chosen based on evaluation metrics.
5. *Cluster Characterization:* The characteristics of each cluster are analyzed using descriptive statistics and visualizations.

## Results

- *Optimal Number of Clusters:* The analysis suggests that 5 clusters (K-Means) and 3 clusters (DBSCAN) are the most appropriate for the dataset.
- *Cluster Characteristics:* Each cluster exhibits distinct characteristics based on the variables used for clustering, such as quantity, value, and weight.

## Managerial Insights

- *Understanding Trade Patterns:* The identified clusters provide insights into different trade activities, such as high-value transactions, bulk commodity trading, and seasonal trends.
- *Targeted Business Strategies:* Businesses can use cluster information to develop targeted strategies for different customer segments, including logistics solutions, pricing models, and product recommendations.
- *Risk Management:* Identifying clusters with unusual or potentially risky characteristics aids in risk mitigation and fraud detection.
- *Operational Efficiency:* Businesses can optimize their operations based on the characteristics of different trade segments, improving inventory management, shipping routes, and customs procedures.

```python
# Calculate variable-wise missing data percentage
variable_missing_percentage = sd48_sample.isnull().sum() * 100 / len(sd48_sample)

# Calculate record-wise missing data percentage
record_missing_percentage = sd48_sample.isnull().sum(axis=1) * 100 / len(sd48_sample.columns)

# Print variable-wise and record-wise missing data information
print("Variable-wise missing data percentage:\n", variable_missing_percentage)
print("\nRecord-wise missing data percentage:\n", record_missing_percentage)

# Drop or impute based on variable missing data percentage
for col in sd48_sample.columns:
  if variable_missing_percentage[col] >= 50:
    # Drop the variable
    sd48_sample = sd48_sample.drop(col, axis=1)
  else:
    # Impute missing values (using mean for numerical, mode for categorical)
    if pd.api.types.is_numeric_dtype(sd48_sample[col]):
      sd48_sample[col] = sd48_sample[col].fillna(sd48_sample[col].mean())
    else:
      sd48_sample[col] = sd48_sample[col].fillna(sd48_sample[col].mode()[0])


# Drop or impute based on record missing data percentage
for index in sd48_sample.index:
  if record_missing_percentage[index] >= 50:
    # Drop the record
    sd48_sample = sd48_sample.drop(index)
  else:
    # Impute missing values (using mean for numerical, mode for categorical)
    for col in sd48_sample.columns:
      if pd.isna(sd48_sample.loc[index,col]):
        if pd.api.types.is_numeric_dtype(sd48_sample[col]):
          sd48_sample.loc[index, col] = sd48_sample[col].mean()
        else:
          sd48_sample.sample[index,col] = sd48_sample[col].mode()[0]
```

```python
def categorical_analysis(df):
    results = []
    for column in df.columns:
        mode_values = df[column].mode()
        mode = mode_values.iloc[0] if not mode_values.empty else np.nan
        frequency = df[column].value_counts().get(mode, 0)



        row = {
            'Variable': column,
            'Count': df[column].count(),
            'Mode': mode,
            'Frequency': frequency,
        }

        # ... (calculations for ordinal variables) ...

        results.append(row)

    return pd.DataFrame(results)

# Apply the function to your DataFrame
analysis_results_df = categorical_analysis(sd48_sample[categorical_cols])

# Display the table
display(analysis_results_df)
```

```python
def plot_categorical_data(df, top_n=10):

    for col in categorical_cols:
        # Count the occurrences of each category
        counts = df[col].value_counts()

        # Check the number of unique values to decide visualization type
        num_unique = len(counts)

        plt.figure(figsize=(10, 6))

        if num_unique <= 10:
            # Pie chart for categories with <= 10 unique values
            plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=90)
            plt.title(f'Pie Chart of {col}')
            plt.axis('equal')  # Equal aspect ratio ensures the pie chart is a circle
        elif num_unique <= 20:
            # Horizontal bar chart for categories with > 10 and <= 20 unique values
            sns.barplot(x=counts.values, y=counts.index, orient='h')
            plt.title(f'Bar Chart of {col}')
            plt.xlabel('Frequency')
            plt.ylabel(col)
        else:
            # Top N categories for larger datasets (>20 unique values)
            top_counts = counts.head(top_n)
            sns.barplot(x=top_counts.values, y=top_counts.index, orient='h')
            plt.title(f'Top {top_n} Bar Chart of {col}')
            plt.xlabel('Frequency')
            plt.ylabel(col)

        plt.show()

# Example usage with your dataset 'sd48_sample'
plot_categorical_data(sd48_sample[categorical_cols])
```

```python
def print_encoding_mapping(df_original, df_encoded):
    """
    Prints the encoding mapping for columns that have been encoded
    using OrdinalEncoder in sd48_sample_encoded.

    Args:
        df_original (pd.DataFrame): The original DataFrame before encoding.
        df_encoded (pd.DataFrame): The encoded DataFrame.
    """

    # Find categorical columns in the original DataFrame
    categorical_cols = df_original.select_dtypes(include=['object']).columns.tolist()

    # Verify if categorical columns are in the encoded dataframe
    categorical_cols = [col for col in categorical_cols if col in df_encoded.columns]

    # Check if the encoding has already been applied
    if len(categorical_cols) > 0:
      # Iterate through the categorical columns in the encoded DataFrame
      for col in categorical_cols:
          print(f"\nEncoding Mapping for {col}:")

          # Get the unique values from the original DataFrame column
          unique_values = sorted(df_original[col].unique())

          # Get the unique encoded values from the encoded DataFrame column
          encoded_values = sorted(df_encoded[col].unique())

          # Check if the unique encoded values are the same as the number of unique original values
          if len(unique_values) == len(encoded_values):
            # Create mapping
            mapping = dict(zip(unique_values, encoded_values))

            # Print the mapping
            for original_value, encoded_value in mapping.items():
                print(f"{original_value} -> {encoded_value}")

          else:
            print(f"Encoding mapping cannot be accurately determined for {col} because the number of unique original values is different from the unique encoded values.")

    else:
      print("No categorical columns found in the original DataFrame, or categorical columns not found in encoded dataframe.")
# Example usage (assuming you have sd48_sample and sd48_sample_encoded):
if 'sd48_sample' in locals() and 'sd48_sample_encoded' in locals():
  print_encoding_mapping(sd48_sample, sd48_sample_encoded)
else:
  print("sd48_sample or sd48_sample_encoded not defined")
```

```python
# Data for clustering - only non-categorical columns
data_for_clustering_k = sd48_sample_encoded[non_categorical_cols].copy()

# Define the range of clusters to test
cluster_range = [2, 3, 4, 5]

# Store results for analysis
results = {}

# Loop through the different numbers of clusters
for n_clusters in cluster_range:
    # Apply K-Means
    kmeans = KMeans(n_clusters=n_clusters, random_state=55048, n_init='auto')
    cluster_labels = kmeans.fit_predict(data_for_clustering_k)

    # Calculate Silhouette Score
    silhouette_avg = silhouette_score(data_for_clustering_k, cluster_labels)

    # Calculate Davies-Bouldin Score
    db_score = davies_bouldin_score(data_for_clustering_k, cluster_labels)

    results[n_clusters] = {
        'silhouette_score': silhouette_avg,
        'davies_bouldin_score': db_score
    }

    # Add cluster labels to the DataFrame
    data_for_clustering_k[f'Cluster_{n_clusters}'] = cluster_labels

    print(f"For n_clusters={n_clusters}:")
    print(f"  Average Silhouette Score: {silhouette_avg}")
    print(f"  Davies-Bouldin Score: {db_score}")

# Plotting the results
plt.figure(figsize=(12, 6))

# Silhouette Score Plot
plt.subplot(1, 2, 1)
silhouette_scores = [results[k]['silhouette_score'] for k in results]
plt.plot(results.keys(), silhouette_scores, marker='o')
plt.title('Silhouette Score for Different Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.xticks(cluster_range)
plt.grid(True)

# Davies-Bouldin Score Plot
plt.subplot(1, 2, 2)
db_scores = [results[k]['davies_bouldin_score'] for k in results]
plt.plot(results.keys(), db_scores, marker='o', color='r')
plt.title('Davies-Bouldin Score for Different Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Davies-Bouldin Score')
plt.xticks(cluster_range)
plt.grid(True)

plt.tight_layout()
plt.show()

# Select the best number of clusters based on Silhouette Score
best_n_clusters_silhouette = max(results, key=lambda k: results[k]['silhouette_score'])
print(f"\nThe optimal number of clusters based on the highest Silhouette Score is: {best_n_clusters_silhouette}")

# Select the best number of clusters based on Davies-Bouldin Score (lower is better)
best_n_clusters_db = min(results, key=lambda k: results[k]['davies_bouldin_score'])
print(f"The optimal number of clusters based on the lowest Davies-Bouldin Score is: {best_n_clusters_db}")
```

```python
# Data for clustering - only non-categorical columns
data_for_clustering = sd48_sample_encoded[non_categorical_cols].copy()

# Define a range of eps values to try
eps_range = np.arange(0.1, 1.1, 0.1)

# Define a range of min_samples values to try
min_samples_range = range(8, 11)

best_silhouette_score = -1
best_db_score = float('inf')
best_eps = None
best_min_samples = None
best_labels = None

# Store results for analysis
results = []

# Loop through the different eps values
for eps in eps_range:
    for min_samples in min_samples_range:
        # Apply DBSCAN
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        cluster_labels = dbscan.fit_predict(data_for_clustering)

        # Check if there is more than one cluster
        if len(set(cluster_labels)) > 1:
            # Calculate Silhouette Score
            silhouette_avg = silhouette_score(data_for_clustering, cluster_labels)

            # Calculate Davies-Bouldin Score
            db_score = davies_bouldin_score(data_for_clustering, cluster_labels)

            #Store results
            results.append({
                'eps': eps,
                'min_samples': min_samples,
                'silhouette_score': silhouette_avg,
                'davies_bouldin_score': db_score
            })

            # Update best_score, if needed
            if silhouette_avg > best_silhouette_score:
                best_silhouette_score = silhouette_avg
                best_db_score = db_score
                best_eps = eps
                best_min_samples = min_samples
                best_labels = cluster_labels
        else:
            # Store results with Nan
            results.append({
                'eps': eps,
                'min_samples': min_samples,
                'silhouette_score': np.nan,
                'davies_bouldin_score': np.nan
            })

# Create the dataframe
results_df = pd.DataFrame(results)

# Print results
print(results_df)

# Print the best values
print(f"\nThe optimal parameters are: eps={best_eps} and min_samples={best_min_samples} (Highest Silhouette Score: {best_silhouette_score}, Lowest DB Score {best_db_score})")

# Add best_labels to the original DataFrame
sd48_sample_encoded['DBSCAN_Cluster'] = best_labels

# Display the number of data points in each cluster
print("\nNumber of data points in each cluster:")
print(sd48_sample_encoded['DBSCAN_Cluster'].value_counts())

# Display the first rows of the DataFrame with cluster labels
print("\nFirst rows of the DataFrame with cluster labels:")
print(sd48_sample_encoded.head())

# Analyze cluster characteristics (e.g., means of variables in each cluster)
print("\nCluster Characteristics (means):")
cluster_means = sd48_sample_encoded.groupby('DBSCAN_Cluster').mean()
print(cluster_means)
```

```python
def evaluate_clustering_models(data):
    """
    Evaluates the performance of 5-Cluster K-Means and DBSCAN clustering models.

    Args:
        data (pd.DataFrame): The dataset for clustering.

    Returns:
        pd.DataFrame: A DataFrame containing the performance metrics for both models.
    """

    # K-Means Clustering with 5 clusters
    kmeans = KMeans(n_clusters=5, random_state=55048, n_init='auto')
    kmeans_labels = kmeans.fit_predict(data)

    # DBSCAN Clustering
    dbscan = DBSCAN(eps=0.073, min_samples=5)
    dbscan_labels = dbscan.fit_predict(data)

    # Performance Metrics for K-Means
    kmeans_silhouette = silhouette_score(data, kmeans_labels)
    kmeans_db_index = davies_bouldin_score(data, kmeans_labels)

    # Performance Metrics for DBSCAN (excluding noise points)
    valid_labels_dbs = dbscan_labels[dbscan_labels != -1]
    valid_data_dbs = data[dbscan_labels != -1]

    if len(valid_labels_dbs) > 1:
        dbscan_silhouette = silhouette_score(valid_data_dbs, valid_labels_dbs)
        dbscan_db_index = davies_bouldin_score(valid_data_dbs, valid_labels_dbs)
    else:
        dbscan_silhouette = np.nan  # Not enough data points to calculate
        dbscan_db_index = np.nan  # Not enough data points to calculate

    # Number of clusters
    n_clusters_kmeans = len(set(kmeans_labels))
    n_clusters_dbscan = len(set(dbscan_labels)) - (1 if -1 in dbscan_labels else 0)

    # Count the noise points in DBSCAN
    n_noise_dbscan = list(dbscan_labels).count(-1)

    # Create a DataFrame for the results
    results = pd.DataFrame({
        'Model': ['K-Means (5 Clusters)', 'DBSCAN'],
        'Number of Clusters': [n_clusters_kmeans, n_clusters_dbscan],
        'Number of Noise Points': [0, n_noise_dbscan],
        'Silhouette Score': [kmeans_silhouette, dbscan_silhouette],
        'Davies-Bouldin Index': [kmeans_db_index, dbscan_db_index]
    })

    return results

data_for_clustering = sd48_sample_encoded[non_categorical_cols].copy()
results_df = evaluate_clustering_models(data_for_clustering)
print(results_df)
```

```python
def k_means_diagnostics(data, cluster_range):
    """
    Performs K-Means clustering diagnostics on the given data.

    Args:
        data (pd.DataFrame): The data to cluster.
        cluster_range (list): A list of the number of clusters to test.

    Returns:
        dict: A dictionary containing the results of the diagnostics.
    """
    results = {}

    for n_clusters in cluster_range:
        kmeans = KMeans(n_clusters=n_clusters, random_state=55048, n_init='auto')
        cluster_labels = kmeans.fit_predict(data)

        silhouette_avg = silhouette_score(data, cluster_labels)
        db_score = davies_bouldin_score(data, cluster_labels)

        results[n_clusters] = {
            'silhouette_score': silhouette_avg,
            'davies_bouldin_score': db_score,
            'cluster_labels': cluster_labels
        }

        print(f"For n_clusters={n_clusters}:")
        print(f"  Average Silhouette Score: {silhouette_avg}")
        print(f"  Davies-Bouldin Score: {db_score}")

    return results

def plot_k_means_diagnostics(results):
    """
    Plots the results of K-Means diagnostics.

    Args:
        results (dict): The results from the k_means_diagnostics function.
    """
    cluster_range = list(results.keys())

    # Plotting the results
    plt.figure(figsize=(12, 6))

    # Silhouette Score Plot
    plt.subplot(1, 2, 1)
    silhouette_scores = [results[k]['silhouette_score'] for k in results]
    plt.plot(cluster_range, silhouette_scores, marker='o')
    plt.title('Silhouette Score for Different Number of Clusters')
    plt.xlabel('Number of Clusters')
    plt.ylabel('Silhouette Score')
    plt.xticks(cluster_range)
    plt.grid(True)

    # Davies-Bouldin Score Plot
    plt.subplot(1, 2, 2)
    db_scores = [results[k]['davies_bouldin_score'] for k in results]
    plt.plot(cluster_range, db_scores, marker='o', color='r')
    plt.title('Davies-Bouldin Score for Different Number of Clusters')
    plt.xlabel('Number of Clusters')
    plt.ylabel('Davies-Bouldin Score')
    plt.xticks(cluster_range)
    plt.grid(True)

    plt.tight_layout()
    plt.show()

    # Select the best number of clusters based on Silhouette Score
    best_n_clusters_silhouette = max(results, key=lambda k: results[k]['silhouette_score'])
    print(f"\nThe optimal number of clusters based on the highest Silhouette Score is: {best_n_clusters_silhouette}")

    # Select the best number of clusters based on Davies-Bouldin Score (lower is better)
    best_n_clusters_db = min(results, key=lambda k: results[k]['davies_bouldin_score'])
    print(f"The optimal number of clusters based on the lowest Davies-Bouldin Score is: {best_n_clusters_db}")

# Assuming sd48_sample_encoded and non_categorical_cols are defined
# and you have a copy of the data ready in data_for_clustering_k
# Define the data and range of clusters to test
if 'sd48_sample_encoded' in locals() and 'non_categorical_cols' in locals():
  data_for_clustering_k = sd48_sample_encoded[non_categorical_cols].copy()
  cluster_range = [2, 3, 4, 5]
  # Run diagnostics
  results = k_means_diagnostics(data_for_clustering_k, cluster_range)
  # Plot the results
  plot_k_means_diagnostics(results)
else:
  print('sd48_sample_encoded and non_categorical_cols variables are not defined, please run previous code')
```

```python
def calculate_cluster_characteristics(data, cluster_labels):
    """
    Calculates cluster characteristics, including homogeneity and heterogeneity.

    Args:
        data (pd.DataFrame): The original DataFrame with non-categorical data.
        cluster_labels (np.ndarray): The cluster labels for each data point.

    Returns:
        pd.DataFrame: A DataFrame containing the homogeneity and heterogeneity scores for each cluster.
    """
    cluster_characteristics = []
    unique_clusters = np.unique(cluster_labels)

    for cluster in unique_clusters:
        cluster_data = data[cluster_labels == cluster]

        # Calculate the mean of each feature for this cluster
        cluster_means = cluster_data.mean()

        # Homogeneity Calculation
        homogeneity_scores = []
        for index, row in cluster_data.iterrows():
            euclidean_distance = np.sqrt(np.sum((row - cluster_means)**2))
            homogeneity_scores.append(euclidean_distance)
        mean_homogeneity_score = np.mean(homogeneity_scores)

        # Heterogeneity Calculation (using standard deviation as a proxy)
        cluster_std = cluster_data.std().mean() # we calculate the mean of the standard deviations of each feature in the cluster.

        cluster_characteristics.append({
            'Cluster': cluster,
            'Mean Homogeneity Score': mean_homogeneity_score,
            'Mean Heterogeneity Score': cluster_std
        })

    return pd.DataFrame(cluster_characteristics)

# Example usage:
if 'sd48_sample_encoded' in locals() and 'non_categorical_cols' in locals():
    # 1. Data and Cluster Labels:
    data_for_clustering_k = sd48_sample_encoded[non_categorical_cols].copy()
    cluster_range = [2, 3, 4, 5]
    # 2. Select number of clusters
    n_clusters = 5  # Best number of clusters selected before

    # 3. Apply K-Means
    kmeans = KMeans(n_clusters=n_clusters, random_state=55048, n_init='auto')
    cluster_labels = kmeans.fit_predict(data_for_clustering_k)

    # 4. Calculate Homogeneity and Heterogeneity:
    characteristics_df = calculate_cluster_characteristics(data_for_clustering_k, cluster_labels)

    # 5. Print the Results:
    print("Cluster Characteristics (Homogeneity and Heterogeneity):")
    print(characteristics_df)
else:
  print("Error: sd48_sample_encoded and non_categorical_cols variables are not defined. Please run previous blocks.")
```
